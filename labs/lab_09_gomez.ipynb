# 2. Normalizar las variables

```python
#Se seleccionan las columnas numéricas:
standard_df = (df
               .select_dtypes(include=['int64','float64'])
               .drop(["diagnosis"],axis=1)
              )
#Se estandariza:
standard_df[standard_df.columns] = StandardScaler().fit_transform(standard_df[standard_df.columns])
standard_df
```

#3. Método de reducción de dimensionalidad 

```python
x = standard_df.values
pca = PCA(n_components=12) 
principalComponents = pca.fit_transform(x) 

percent_variance = np.round(pca.explained_variance_ratio_* 100, decimals =2)
columns = [f"PC{i}" for i in range(1,13)]

plt.figure(figsize=(20,10))
plt.bar(x= range(1,13), height=percent_variance, tick_label=columns)
plt.ylabel('Percentate of Variance Explained')
plt.xlabel('Principal Component')
plt.title('PCA Scree Plot')
plt.show()
```

```python
percent_variance_cum = np.cumsum(percent_variance)

columns_sum =[f"PC1+...+PC{i+1}" for i in range(2,12)]
columns_sum = ["PC1", "PC1+PC2"] + columns_sum

plt.figure(figsize=(20,10))
plt.bar(x= range(1,13), height=percent_variance_cum, tick_label=columns_sum)
plt.ylabel('Percentate of Variance Explained')
plt.xlabel('Principal Component Cumsum')
plt.title('PCA Scree Plot')
plt.show()
```

##4.1 Modelo de Arbol de clasificación

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier,DecisionTreeRegressor
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn import tree
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(principalComponents, df['diagnosis'], test_size=0.3, random_state = 42)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, recall_score, precision_score, f1_score
def summary_metrics(y_pred,model_pred):
    """
    Aplicar las distintas metricas definidas
    :param y_pred,model_pred: pandas series del mismo largo a calcular metricas
    :return: No se retorna nada pero se printean las metricas
    """
    print('\nMatriz de confusion:\n ')
    print(confusion_matrix(y_pred,model_pred))
    
    # metrics
    
    print('\nMetricas:\n ')
    print('accuracy:   ',accuracy_score(y_pred, model_pred))
    print('recall:     ',recall_score(y_pred, model_pred))
    print('precision:  ',precision_score(y_pred, model_pred))
    print('f-score:    ',f1_score(y_pred, model_pred))
    print("")

    return
```

```python
# creación del modelo
model = DecisionTreeClassifier()

# rango de parametros
rango_criterion = ['gini','entropy']
rango_max_depth =np.array( [4,5,6,7,8,9,10,11,12,15,20,30,40,50,70,90,120,150])
param_grid = dict(criterion=rango_criterion, max_depth=rango_max_depth)
param_grid
```

```python
# aplicar greed search
gs = GridSearchCV(estimator=model, 
                  param_grid=param_grid, 
                  scoring='accuracy',
                  cv=5,
                  n_jobs=-1)

gs = gs.fit(X_train, Y_train)
```

```python
# imprimir resultados
print(gs.best_score_)
print(gs.best_params_)
```

```python
# utilizando el mejor modelo
mejor_modelo = gs.best_estimator_
mejor_modelo.fit(X_train, Y_train)
print('Precisión: {0:.3f}'.format(mejor_modelo.score(X_test, Y_test)))
```

```python
dt_pred= mejor_modelo.predict(X_test)
```

```python
summary_metrics(Y_test, dt_pred)
```
